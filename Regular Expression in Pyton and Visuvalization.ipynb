```python
#add text on the bar
ax = d.plot(kind = "bar", color = ['blue','red'])
for i in ax.patches:
  ax.text(i.get_x()+i.get_width(), #x
          i.get_height()+1, #y
          str(int(i.get_height()))) #s

plt.title("Bar plot of Attrition")
plt.show()
```
