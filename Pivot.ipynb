```python
import pandas as pd
#sample dataset
df = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two',
                          'two'],
                  'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                  'baz': [1, 2, 3, 4, 5, 6],
                  'zoo': ['x', 'y', 'z', 'q', 'w', 't']})
df
```

```python
wide_data = df.pivot(index='foo', columns='bar', values='baz')
```

```python
wide_data.unstack().reset_index()
```
